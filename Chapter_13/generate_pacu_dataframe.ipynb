```python
import pandas as pd
import numpy as np

import random, string
```

# Generate PACU DataFrame

```python
def generate_mrn_df():
    mrn = ''.join(random.choices(string.ascii_letters + string.digits, k=6)).upper()
    dc_loc = random.sample(['Floor 10','Floor 5','Floor 4'], 1)[0]
    num_obs = random.sample([8,9,10], 1)[0]
    freq = random.sample(['D','D','h'], 1)[0] 

    pacu_df = pd.DataFrame(data={
        'Patient_ID': mrn,
        'InternalTemperature':[round(i,1) for i in np.random.normal(loc=37.0, scale=0.75, size=num_obs)],
        'SurfaceTemperature':[round(i,1) for i in np.random.normal(loc=36.1, scale=0.75, size=num_obs)],
        'OxygenSat':[i if i <= 100 else 100 for i in np.random.normal(loc=97, scale=5, size=num_obs).astype(int)],
        'BloodPressure':[f"{sys}/{dias}" for sys,dias in zip(np.random.normal(loc=120, scale=5, size=num_obs).astype(int),
                        np.random.normal(loc=80, scale=5, size=num_obs).astype(int))],
        'ComfortLevel ':num_obs * random.sample(['poor ','adequate ','good '], 1)[0].split(' ')[:-1],
        'Vitals_RecordTime':pd.date_range(start=pd.to_datetime('2019-10-11'), periods=num_obs, freq=freq),
        'DischargeLocation':dc_loc
    })
    
    return(pacu_df)
```

```python
num_pts = 20
all_patients_raw = pd.concat([generate_mrn_df() for i in range(num_pts)])
```

```python
for col in ['InternalTemperature','OxygenSat','ComfortLevel ']:
    if col == 'ComfortLevel ':
        all_patients_raw[col] = all_patients_raw[col].mask(np.random.random(len(all_patients_raw)) < .2)
    else:
        all_patients_raw[col] = all_patients_raw[col].mask(np.random.random(len(all_patients_raw)) < .1)
```
