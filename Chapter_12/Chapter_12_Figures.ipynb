```python
import pandas as pd

import numpy as np

from matplotlib import cm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x=[1,2,3,4], y=[1,2,2,3])
plt.show()
```

```python
coords_df = pd.DataFrame(data={'label':['a','a','b','b'],'x_coords':[1,2,3,4], 'y_coords':[1,2,2,3]})

coords_df['label'] = coords_df['label'].astype("category")

plt.scatter(x=coords_df['x_coords'], y=coords_df['y_coords'])
plt.show()
```

```python
viridis_map = cm.get_cmap('viridis', 2)

plt.figure(figsize=(12,6))
plt.scatter(x=coords_df['x_coords'], y=coords_df['y_coords'], c=viridis_map(coords_df['label'].cat.codes))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.scatter(x=coords_df['x_coords'], y=coords_df['y_coords'])
ax.set_xlabel("Some x-axis label", fontsize=24)
plt.show()
```
